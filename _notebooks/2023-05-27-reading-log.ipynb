```python
import sqlite3
from flask import Flask, request, render_template

app = Flask(__name__)

def make_database():
    # Connect to the database (will create it if it doesn't exist)
    connection = sqlite3.connect('instance/readinglog.db')
    cursor = connection.cursor()

    # Create the readinglog table if it doesn't already exist
    cursor.execute('''CREATE TABLE IF NOT EXISTS readinglog (
                    username TEXT,
                    name TEXT,
                    start_date TEXT,
                    finish_date TEXT,
                    book TEXT,
                    rating REAL
                )''')

    # Commit changes and close the connection
    connection.commit()
    connection.close()


make_database()


@app.route('/', methods=['GET'])
def index():
    return render_template('index.html')


@app.route('/create', methods=['POST'])
def create():
    username = request.form['username']
    name = request.form['name']
    start_date = request.form['start_date']
    finish_date = request.form['finish_date']
    book = request.form['book']
    rating = request.form['rating']

    # Connect to the database and create a cursor to execute SQL commands
    connection = sqlite3.connect('instance/readinglog.db')
    cursor = connection.cursor()

    try:
        # Execute SQL to insert record into db
        cursor.execute("INSERT INTO readinglog (username, name, start_date, finish_date, book, rating) VALUES (?, ?, ?, ?, ?, ?)",
                       (username, name, start_date, finish_date, book, rating))
        # Commit the changes
        connection.commit()

        return "Success"

    except sqlite3.Error as error:
        return "Error: " + str(error)

    finally:
        # Close cursor and connection
        cursor.close()
        connection.close()


@app.route('/read', methods=['GET'])
def read():
    try:
        # Open a connection to the database and create a cursor
        connection = sqlite3.connect('instance/readinglog.db')
        cursor = connection.cursor()

        # Fetch all records from the readinglog table
        cursor.execute("SELECT * FROM readinglog")
        rows = cursor.fetchall()

        # If there are any records, return them
        if len(rows) > 0:
            return render_template('read.html', rows=rows)
        else:
            return "There are no books in the list."

    except sqlite3.Error as error:
        return "Error: " + str(error)

    finally:
        # Close the cursor and the connection to the database
        cursor.close()
        connection.close()


@app.route('/update', methods=['POST'])
def update():
    book = request.form['book']
    start_date = request.form['start_date']
    finish_date = request.form['finish_date']
    new_book = request.form['new_book']
    rating = request.form['rating']

    # Connect to the database and create a cursor
    connection = sqlite3.connect('instance/readinglog.db')
    cursor = connection.cursor()

    try:
        # Retrieve the current record from the database
        cursor.execute("SELECT * FROM readinglog WHERE book=?", (book,))
        record = cursor.fetchone()

        # If the book is found, update the record
        if record:
            # Execute SQL to update the record
            cursor.execute("UPDATE readinglog SET start_date=?, finish_date=?, book=?, rating=? WHERE book=?",
                           (start_date, finish_date, new_book, rating, book))
            connection.commit()

            return "Success"

        # If the book is not found, return an error message
        else:
            return "Error: Book not found."

    except sqlite3.Error as error:
        return "Error: " + str(error)

    finally:
        # Close cursor and connection
        cursor.close()
        connection.close()


@app.route('/delete', methods=['POST'])
def delete():
    # Connect to the database and create a cursor
    connection = sqlite3.connect('instance/readinglog.db')
    cursor = connection.cursor()

    # Get the book's name to delete
    book = request.form['book']

    try:
        # Use a SQL query to find the book with the given name
        cursor.execute("SELECT * FROM readinglog WHERE book=?", (book,))
        row = cursor.fetchone()

        # If the book exists, delete the record
        if row:
            cursor.execute("DELETE FROM readinglog WHERE book=?", (book,))
            connection.commit()
            return "Success"

        # If the book is not found, return an error message
        else:
            return "Error: Book not found."

    except sqlite3.Error as error:
        return "Error: " + str(error)

    finally:
        # Close the cursor and the connection to the database
        cursor.close()
        connection.close()


if __name__ == '__main__':
    app.run()
```
